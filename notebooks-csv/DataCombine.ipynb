```python
def drop_rows(df, website_name, drop_size, random_state=42):
    drop_indices = df[df['website'] == website_name].sample(
        n=drop_size, random_state=random_state
    )
    return df.drop(drop_indices.index), drop_indices
```

```python
len(CombinedHumanDF) - len(CombinedAIDF)
```

```python
CombinedHumanDF, HumanUnusedDF = drop_rows(CombinedHumanDF,'https://codeforces.com',1012075) # remove codeforces source codes.
```

```python
def assign_dataset_type(df, train_ratio=0.8, valid_ratio=0.1, test_ratio=0.1, random_state=42):
    dfs = []
    for website, group in df.groupby('website'):
        group = group.sample(frac=1, random_state=random_state).reset_index(drop=True)
        n = len(group)
        n_train = int(n * train_ratio)
        n_valid = int(n * valid_ratio)
        group.loc[:n_train-1, 'dataset_type'] = 'train'
        group.loc[n_train:n_train+n_valid-1, 'dataset_type'] = 'validation'
        group.loc[n_train+n_valid:, 'dataset_type'] = 'test'
        dfs.append(group)
    return pd.concat(dfs).reset_index(drop=True)
```

```python
SplitedHumanDF = assign_dataset_type(CombinedHumanDF)
SplitedHumanDF
```

```python
SplitedAIDF = assign_dataset_type(CombinedAIDF)
SplitedAIDF
```
