## Source codes (Human)

```python
source_code_path = "./Humans/Cleaned/"
df = pd.DataFrame()

HumanDFs = []

# for p in os.listdir(source_code_path):
#     if p.endswith(".csv"):
#         df = pd.concat([df, pd.read_csv(source_code_path + p)])

for p in os.listdir(source_code_path):
    if p.endswith(".csv"):
        HumanDFs.append(pd.read_csv(source_code_path + p))
```

```python
CombinedHumanDF = pd.DataFrame()
for df in HumanDFs:
    CombinedHumanDF = pd.concat([CombinedHumanDF,df])
CombinedHumanDF
```

```python
CombinedHumanDF['submission_id'] = pd.to_numeric(CombinedHumanDF['submission_id'], errors='coerce').astype('Int64')
CombinedHumanDF
```

## Final Split

```python
test_df = pd.read_csv('../Splits/combined/test.csv')
train_df = pd.read_csv('../Splits/combined/train.csv')
validation_df = pd.read_csv('../Splits/combined/validation.csv')
sub_test_df = pd.read_csv('../Splits/combined/validation.csv')

harder_test_df = pd.read_csv('../Splits/harder_test/test.csv')
harder_sub_test_df = pd.read_csv('../Splits/harder_test/sub_test.csv')
```

```python
test_df = test_df.drop(columns=['submission_id'])
train_df = train_df.drop(columns=['submission_id'])
validation_df = validation_df.drop(columns=['submission_id'])
harder_test_df = harder_test_df.drop(columns=['submission_id'])
harder_sub_test_df = harder_sub_test_df.drop(columns=['submission_id'])
```

```python
test_df.to_csv('../Splits/Public/test.csv',index=False)
train_df.to_csv('../Splits/Public/train.csv',index=False)
validation_df.to_csv('../Splits/Public/validation.csv',index=False)
sub_test_df.to_csv('../Splits/Public/validation.csv',index=False)
harder_test_df.to_csv('../Splits/Public/Harder/test.csv',index=False)
harder_sub_test_df.to_csv('../Splits/Public/Harder/sub_test.csv',index=False)
```
