```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import asyncio
import time
from concurrent.futures import ThreadPoolExecutor
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# Configure Selenium WebDriver
def setup_driver(silent=False):
    options = Options()
    if silent:
        options.add_argument("--headless")
    service = Service()
    driver = webdriver.Chrome(service=service, options=options)
    return driver
```

```python
import pandas as pd
```

```python
# https://programming.in.th/submissions/100000 First Submission, need to run for 2e5 (I don't know if server could handle this)
# Last submission is 336798
test_driver = setup_driver()
test_driver.get("https://programming.in.th/tasks")
```

```python
elements = test_driver.find_elements(By.XPATH,'//a[@class="flex w-full rounded-xl px-6 py-3 font-display shadow-md transition group-hover:shadow-lg dark:bg-slate-700"]')
```

```python
# Solved Count
elements[0].find_element(By.XPATH, './/p[@class="text-sm text-gray-500 dark:text-gray-200"]').text
```

```python
# Main Problem ID
elements[0].get_attribute("href")
```

```python
def problem_id_toPDFurl(id):
    return f"https://programming.in.th/api/tasks/{id}/statement"

def problem_id_toSubmission(id):
    return f"https://programming.in.th/tasks/{id}/submissions"
```

```python
lis = []
for element in elements:
    problem_id = element.get_attribute("href").split('/')[-1]
    submissions = int(element.find_element(By.XPATH, './/p[@class="text-sm text-gray-500 dark:text-gray-200"]').text)
    lis.append((problem_id,problem_id_toPDFurl(problem_id),problem_id_toSubmission(problem_id),submissions))
```

```python
lis
```

```python
df = pd.DataFrame(lis, columns=["ProblemID", "PdfURL", "SubmissionURL", "submissions_count"])
df
```
