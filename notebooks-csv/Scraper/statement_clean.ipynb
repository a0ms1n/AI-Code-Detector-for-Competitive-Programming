```python
import pandas as pd
```

# Grader Chan

```python
df = pd.read_csv('./graderchan.csv')
df['ProblemURL'][0]
```

```python
def convert_problem_to_submission_url(problem_url):
    problem_name = problem_url.split('/')[-2]
    submission_url = f"https://firefly.gchan.moe/submission?per_page=100&problemset={problem_name}"
    return submission_url

def convert_problem_to_id(problem_url):
    problem_name = problem_url.split('/')[-2]
    return problem_name
```

```python
df['ProblemSubmissionURL'] = df['ProblemURL'].apply(convert_problem_to_submission_url)
df['problemID'] = df['ProblemURL'].apply(convert_problem_to_id)
df
```

```python
df.to_csv('./StatementCleaned/graderchan_clean.csv',index=False)
```
